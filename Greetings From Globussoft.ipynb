# Question: - 2

One rich person decided to make himself a great garden. The garden should have a from of 
equilateral triangle. There should be a gazebo inside the garden. The gazebo will be connected 
with the triangle vertexes by roads. The lengths of all three roads are known. Those numbers are 
sacred for this rich man. The expense of building such a garden will depend on the area of the 
garden. Help the rich man by calculating what will be the area of his garden.

The first line of the input contains number t â€“ the amount of tests. Then t test descriptions follow. 
Each test consist of three integers a, b, c - the lengths of the roads. It is guaranteed that it's 
possible to build such a garden.


# SOLUTION:-

Let's consider the picture above would be a rich man's garden. Where "D" is gazebo and "AD", "BD", "CD" are three roads.     
Let, length of AD= d, length of BD= e and length of CD = f

Now we have a formula for calculating the area of an equilateral triangle *(I can prove this formula)

Where S= The Area Of The Triangle and s= (d+e+f)/2

```python
#variable assigned for Count the number of test
test_number=1
#variable assigned for input the length of AD
d=int(input("PLEASE INPUT THE LENGTH OF AD ="))
#variable assigned for input the length of BD
e=int(input("PLEASE INPUT THE LENGTH OF BD ="))
#variable assigned for input the length of CD
f=int(input("PLEASE INPUT THE LENGTH OF CD ="))
```

```python
#import math for sqrt function
import math
```

```python
#DIVIDING THE MAIN FORMULA INTO SMALL SMALL PART
p=d**2*(math.sqrt(3)/4)
q=e**2*(math.sqrt(3)/4)
r=f**2*(math.sqrt(3)/4)
s=(d+e+f)/2
x=3*math.sqrt(s*(s-d)*(s-e)*(s-f))
ans=(p+q+r+x)/2
```

```python
#Final Output
print("Test Number=", test_number)
print("The Area of the Traingle=",round(ans,2))
#for the next experiment
test_number+=1
```
